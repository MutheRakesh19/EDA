```python
categorical_columns = ['ID', 'ALIGN', 'EYE', 'HAIR', 'SEX', 'GSM', 'ALIVE']
for col in categorical_columns:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
numerical_columns = ['APPEARANCES', 'YEAR']
for col in numerical_columns:
    df[col].fillna(df[col].median(), inplace=True)
```

```python
# cat vs num
def bi_var_num(col):
    fig,ax=plt.subplots(ncols=2,nrows=1,figsize=(12,4))
    sns.boxplot(x=col,y='ALIGN',data=df,ax=ax[0])
    sns.boxplot(x=col,y='ALIGN',data=df,ax=ax[1])
num_cols=(['page_id', 'APPEARANCES', 'YEAR'])
for i in num_cols:
    bi_var_num(i)
    
```
